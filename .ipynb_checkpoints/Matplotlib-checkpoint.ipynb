```python
## Plotting using matplotlib

## plt scatter (scatter plot)

plt.scatter(x,y,c='g') # 's'- solid value, 'c'- color
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')

# for saving the image 
plt.savefig('Test.png')
```

```python
## plt plot
# check parameter
y=x*x
plt.plot(x,y,'ro--',linestyle='dashed',linewidth=2, markersize=12)
#'r--' multiple dash
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("plt plot")
```

```python
# subplots(within 1 canvas , i can create multiple subplots)

plt.subplot(2,2,1) # (2,2,1)- no of rows , no of column , position- index
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')
# 2 row , 2 column i.e. only 2 position
```

```python
# performing complicated task on array
q=np.arange(1,11)
w=3*q+5
plt.title("demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(q,w)
```

```python
# sin wave diagram

e=np.arange(0,4*np.pi,0.1) #(start, end, step size)
r=np.sin(e)
plt.plot(e,r)
```

```python
# subplot()
z=np.arange(0,5*np.pi,0.1) #(start, end, step size)
c=np.sin(z)
v=np.cos(z)

# set up subplotof height 2 and width 1

#make the first plot
plt.subplot(2,1,1)
plt.plot(z,c,'b')
plt.title('sine')

#make the second plot
plt.subplot(2,1,2)
plt.plot(z,v,'g')
plt.title('cos')
```

```python
## Bar Graph

a=[2,8,10]
b=[11,16,9]

a2=[3,9,11]
b2=[6,15,7]
plt.bar(a,b,color='r')
plt.bar(a2,b2,color='g')
plt.title('Bar Graph')
plt.ylabel('y axis')
plt.xlabel('x axis')
#compare
```

```python
# Histograms
h=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(h,bins=50)
#bins= width or distance of the bar plot
# shift + tab  above
plt.title("histogram")
# It helps us to find out the numbers , what is the density/count we can see on the y axis
# x- all the values
# y - density or the count
# no of vale fall in 20 range i.e. 2
```

```python
# Box plot ( find percentile)
data=[np.random.normal(0,std,100) for std in range(1,4)]
# normal- normal distribution
#loop(0-3)
#data will be list because of for loop
#3 different list will be created and each list will have 3 different array

# Rectangule box plot
plt.boxplot(data,vert=True, patch_artist=True);
#"vert"- vertical
#"patch_artist"-if it is false , the color will get empty
# start from down the counting in graph
# circle are outliers
```

```python
## Pie chart
labels='python','c++','ruby','Java'
sizes=[215,130,245,210] # it will be calculated based on percentage
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) # explode first slice
#explode- how much first slide needs to go

# plot
plt.pie(sizes, explode=explode , labels=labels , colors=colors ,autopct='%1.1f%%', shadow=True)
#autopct- auto point- in what format i want , f- float format, (1.1)- 1 point decimal
# shadow - for slice
plt.axis('equal')
plt.show()
```
