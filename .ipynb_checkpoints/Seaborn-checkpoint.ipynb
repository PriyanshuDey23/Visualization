# If I have 2 features(2d) , It is called univariate analysis
# If I have 3 features(3d), It is called bi variate analysis

```python
# It will help us to analyse multiple features in our dataset
# Different Types of Plot:-
#displot :- 
#jointplot :- It will help in bivariate analysis
#pairplot :- more than 2 independent features , it in real world we cannnot see 4d plot, in that case we use pair plot

import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')

# we have to create a model that will help us to find the 'tip' based on the other features
```

```python
df # restraurant dataset
# tip is dependent feature
```

```python
#co-relation, how related they are(the value should be float or int)
df.corr() # only 3 features based on datatypes
# we are getting the corelation percentage 
# tip and total bill are 67% coreletion
# It means total bill and tip are increasing(+)
# It means total bill and tip are decreasing(-)
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr()) # insering the matrix of coreletion
#heatmap- Plot rectangular data as a color-encoded matrix
# highly corelated is 0.1 and less corelated is 0.5
```

# JoinPlot
# a join plot allows to study the relationship between 2 numeric variables. it help us to do univariate analysis

```python
# join plot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#kind=hex - instead of point , it will show hexagonal shape
# histogram will be shown (both for histogram for both feature)
# maximum tip is around 2 dollar
# maximum bill is around 10 to 20
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
# reg means regression
```

```python
# more than 2 independent feature we use
# pairplot
sns.pairplot(df)
# It will do the permudation and combination of features 
```

```python
# now i may consider different features so that we can classify that point
sns.pairplot(df,hue='sex')
# try with other features
```

```python
# Distplot- It help in creating histogram
sns.distplot(df['tip'])
# we are getting in percentage
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```

# Categorical Plots
# boxplot , violinplot , countplot , bar plot

# It will  basically work  for categorical features

```python
##Countplot:- It basically indicates Show the counts of observations in each categorical bin using bars.

sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# bar plot ( we have to mention our x and y axis value)
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
# boxplot

sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(orient='v', data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue="smoker",data=df)
#hue = classify the x and y with respect to smoker
```

```python
# violin Plot :- Violin plot helps helps us to see both the distribution of the data in terms of kernel density estimation and the box plot

sns.violinplot(x='total_bill',y='day',data=df)
# internel is boxplot
# also shows the distribution
```
